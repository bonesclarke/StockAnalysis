```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
import pandas_datareader.data as web
import datetime as dt
import plotly.express as px
import plotly.graph_objects as go
```

```python
# set dates, form input
start = dt.datetime(2021,1,14)
end = dt.datetime.now()
```

```python
# retrieve stock data
CLNE = web.DataReader(['CLNE'], 'yahoo', start, end)
MT = web.DataReader(['MT'], 'yahoo', start, end)
MT_close = pd.DataFrame(web.DataReader(['MT'], 'yahoo', start, end)['Close'])

# select stock, create a form input using django
SYM = MT
SYM_close = MT_close
```

```python
# set up candlesticks with moving average lines
SYM_trace1 = {
    'x' : SYM_close.index, 
    'open' : SYM[('Open',  'MT')],
    'high' : SYM[('High',  'MT')],
    'low' : SYM[('Low',  'MT')], 
    'close' : SYM[('Close',   'MT')],
    'type' : 'candlestick',
    'name': 'MT'
}

## moving averages
SYM_20 = SYM_close['MT'].rolling(window=20, min_periods=1).mean()
SYM_50 = SYM_close['MT'].rolling(window=50, min_periods=1).mean()
SYM_200 = SYM_close['MT'].rolling(window=200, min_periods=1).mean()

SYM_trace2 = {
    'x': SYM_close.index,
    'y': SYM_20,
    'type': 'scatter',
    'mode': 'lines',
    'line': {
        'width': 1,
        'color': 'blue'
            },
    'name': 'Moving Average of 20 periods'
}

SYM_trace3 = {
    'x': SYM_close.index,
    'y': SYM_50,
    'type': 'scatter',
    'mode': 'lines',
    'line': {
        'width': 1,
        'color': 'red'
    },
    'name': 'Moving Average of 50 periods'
}

SYM_trace4 = {
    'x': SYM_close.index,
    'y': SYM_200,
    'type': 'scatter',
    'mode': 'lines',
    'line': {
        'width': 1,
        'color': 'yellow'
    },
    'name': 'Moving Average of 200 periods'
}

SYM_trace5 = {
    'x': SYM_close.index,
    'y': SYM_200,
    'type': 'scatter',
    'mode': 'lines',
    'line': {
        'width': 1,
        'color': 'yellow'
    },
    'name': 'Moving Average of 200 periods'
}

data = [SYM_trace1, SYM_trace2, SYM_trace3, SYM_trace4, SYM_trace5]

SYM_candlestick = go.Figure(data = data)

SYM_candlestick.update_xaxes(
    title_text = 'Date',
    rangeslider_visible = False,
    rangeselector = dict(
        buttons = list([
            dict(count = 1, label = '1M', step = 'month', stepmode = 'backward'),
            dict(count = 6, label = '6M', step = 'month', stepmode = 'backward'),
            dict(count = 1, label = 'YTD', step = 'year', stepmode = 'todate'),
            dict(count = 1, label = '1Y', step = 'year', stepmode = 'backward'),
            dict(step = 'all')])))

SYM_candlestick.update_layout(
    title = {
        'text': 'MT Share Price',
        'y':0.9,
        'x':0.5,
        'xanchor': 'left',
        'yanchor': 'top'})

SYM_candlestick.update_yaxes(title_text = 'Share Price', tickprefix = '$')
SYM_candlestick.show()
```
